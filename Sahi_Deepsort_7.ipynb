```python
import cv2
import torch
import numpy as np
from ultralytics import YOLO
from sahi import AutoDetectionModel
from sahi.predict import get_sliced_prediction
from deep_sort_realtime.deepsort_tracker import DeepSort
import supervision as sv

# =====================
# CONFIG
# =====================
VIDEO_PATH = "/content/Trial5/vid1.mp4"   # your video file
OUTPUT_PATH = "/content/output_sahi_deepsort3.mp4"
MODEL_PATH = "/content/Trial5/best.pt"  # replace with your trained model

# =====================
# LOAD MODELS
# =====================
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the YOLO model directly
yolo_model = YOLO(MODEL_PATH)

# Use the loaded YOLO model with SAHI
detection_model = AutoDetectionModel.from_pretrained(
    model_type="yolov8", # Specify yolov8 as model type if using YOLOv8
    model=yolo_model,
    confidence_threshold=0.7,
    device=device,
)


tracker = DeepSort(
    max_age=30,
    n_init=2,
    nms_max_overlap=1.0,
    max_cosine_distance=0.3,
    nn_budget=None,
    override_track_class=None,
    embedder="mobilenet",
    half=True,
    bgr=True,
    embedder_gpu=True,
)

# =====================
# VIDEO I/O
# =====================
cap = cv2.VideoCapture(VIDEO_PATH)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

out = cv2.VideoWriter(OUTPUT_PATH, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))


# =====================
# RUN DETECTION + TRACKING
# =====================
frame_idx = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frame_idx += 1
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # SAHI sliced prediction
    result = get_sliced_prediction(
        rgb,
        detection_model,
        slice_height=256,
        slice_width=256,
        overlap_height_ratio=0.2,
        overlap_width_ratio=0.2,
    )

    # Prepare detections for DeepSORT
    detections_for_deepsort = []
    for pred in result.object_prediction_list:
        x1, y1, x2, y2 = map(int, pred.bbox.to_xyxy())
        bbox_xywh = (x1, y1, x2 - x1, y2 - y1) # xywh format
        confidence = pred.score.value
        class_id = pred.category.id if pred.category else 0
        detections_for_deepsort.append((bbox_xywh, confidence, class_id))


    # Update tracker
    tracks = tracker.update_tracks(detections_for_deepsort, frame=frame)

    labels = []
    boxes = []
    for track in tracks:
        if not track.is_confirmed():
            continue
        x1, y1, x2, y2 = track.to_ltrb()
        boxes.append([x1, y1, x2, y2])
        labels.append(f"ID {track.track_id}")

    if len(boxes) > 0:
        detections_sv = sv.Detections(
            xyxy=np.array(boxes),
            confidence=np.ones(len(boxes)),
            class_id=np.zeros(len(boxes), dtype=int), # Ensure class_id is integer type
        )
        frame = box_annotator.annotate(scene=frame, detections=detections_sv)
        frame = label_annotator.annotate(scene=frame, detections=detections_sv, labels=labels)

    out.write(frame)
    if frame_idx % 50 == 0:
        print(f"Processed {frame_idx} frames")

cap.release()
out.release()
print(f"Saved result video to {OUTPUT_PATH}")
```
